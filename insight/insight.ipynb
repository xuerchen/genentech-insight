## Introduction

####This study is done by Jiwei Liu and Xueer Chen, part of the team better luck in 2016, 6th place of the Genentech Cervical Cancer Screening contest. We first study the implact of leaks, absence of a claim line in either diagnosis_head or procedure_head in the datasets, and show that their predicting power is dominant. Therefore in the following analysis we remove all the records with leak features to better identify other important features. We modified Follow the Regularized Leader (FTRL) algorithm to extract important features and feature interactions in a online fashion. Our approach can find both leak and good features automatically without any prior domain knowledge.


###In summary, we make the following insights:
####1. The leak is frequent and strong. 
<font color ='red'> 31.8% </font> Patients have leak features and <font color ='red'> 97.8%</font>of them are screeners. So we remove the records with leak to do the following analysis.
####2. Non-leak features are also very useful: diagnosis code, procedure code, surgical code and their interactions.
Dignosis code and procedure code pairs linked by the same claim ID can be a strong indicator of screeners. Specifically, the algorithm found <font color ='red'>109</font> code pairs that correspond to more than <font color ='red'>90%</font> screener rate, <font color ='red'>41</font> of which correspond to <font color ='red'>+99%</font> screener rate.
 
A single diagnosis code, procedure code or surgical code is also a good indicator of screeners but much more frequent in data sets. The algorithm found <font color ='red'>27</font> such single codes that correspond to more than 80% screener rate, with frequency (percentage of patients who have these features) ranging from <font color ='red'>1%</font> to <font color ='red'>14%</font>

A single procedure code or a code pair can be a good indicator of non-screeners. The algorithm found <font color ='red'>4</font> codes and <font color ='red'>5</font> pairs which correspond to less than <font color ='red'>30%</font> screener rate and more than <font color ='red'>0.01%</font> frequency. The algorithm also found <font color ='red'>34</font> such codes and pairs that correspond to <font color ='red'>0</font> screener rate but they are extremely infrequent.

####3. Most useful non-code features are primary_practitioner_id and specialty_code
####4. Some practitioners' job directly relates to screening or cervical cancer.
For example, practitioner <font color ='red'>"16444459", "12709441", and "14015592"</font> are doing DIAGNOSTIC RADIOLOGY and OBSTETRICS AND GYNECOLOGY and they correspond to <font color ='red'>+90%</font> screener rate!
####5. Nurse's experience matters
For example, NMW, NURSE MIDWIFE, advanced practice registered nurses, correspond to <font color ='red'>81%</font> screener rate while SNP, SCHOOL NURSE PRACTITIONER, correspond to only <font color ='red'>25%</font> screener rate

## A closer look at leak


###We extract 400 most important features using ftrl from diagnosis_head.csv and procedure_head.csv, which only include occurrences of diagnosis codes, procedure codes and their interactions. With these 400 features, a ftrl model can get 0.908 auc in less than 10 minutes. 

###The features can be divided into three categrories: 
1. the occurrence of a single diagnosis/prucedure code
2. the occurrence of a pair of a diagnosis code and a prucedure code connected by the same claim id.
3. the occurrence of a pair of a diagnosis code and a missing prucedure code connected by the same claim id. 

###As we know, type 3) the missing code or entire procedure entries are actually leaks. 


```python
fea=pd.read_csv('../online2/feamap8.csv').sort(columns='frequency',axis=0,ascending=False)
fea['is_leak']=fea.feature.map(str).apply(lambda x: x[-1]=='_').astype(int)
fea['is_positive']=fea.weight>0
fea[:10]
```

feamap8.csv is the feature selection result from ftrl which include important codes and codes interactions. All original occurrences of codes and their interactions are mapped/hashed to 2^28 feature space. 

1. The column 'index' indicates the feature's index in the 2^28 feature space. 
2. The column 'feature' indicates the content of the feature, from which we can tell the 3 types. For example, V77.71 is the 1st type or occurrence of a single diagnosis code. 'V76.41_G0328' is the 2nd type or a pair of codes. V76.41 is a diagnosis code and G0328 is a procedure code and they are connected by the same claim_id. 'V88.01_' is the 3rd type or a leak feature. V88.01 is a diagnosis code but the corresponding procedure code with the same claim_id is missing. 
3. The column 'weight' stands for the feature weight in ftrl model. A positive/negative weight indicates a screener/non-screener accordingly.
4. The column 'screen_rate' indicates the percentage of screeners in patients whose records include this feature at least once.
5. The column 'frequency' indicates the percentage of patients whose records include this feature at least once out of all the patients
6. The column 'description' includes the descriptions of the codes. For code pairs descriptions are in the format of "diagnosis code description |||| procedure code description"
7. The column 'is_leak' indicates wheather the feature is a leak feature. (diagnosis_code + missing procedure code of a claim)
8. The column 'is_positive' indicates if a feature has a positive weight

```python
from matplotlib import font_manager as fm
proptease = fm.FontProperties()
proptease.set_size('x-large')
proptease.set_weight('bold')
fig = plt.figure(figsize=(15,5))

import csv
a,b,c,d=0,0,0,0
for row in csv.DictReader(open('../online2/features/fea8.csv')):
    flag=0
    for x in row['features'].split():
        if x[-1]=='_':
            flag=1
            break
    if flag:
        if float(row['is_screener'])>0:
            a+=1
        else:
            b+=1
    else:
        if float(row['is_screener'])>0:
            c+=1
        else:
            d+=1
ax3 = fig.add_subplot(121)
labels = 'Screeners with leak featuers', 'Non-Screeners with leak featuers','Screeners without leak features','Non-Screeners without leak features'
sizes = a,b,c,d
colors = ['yellowgreen', 'lightcoral','c','m']
explode = ( 0.1, 0.12,0,0)  
patches, texts3, autotexts3 = ax3.pie(sizes,explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
ax3.set_title('Patiensts breakdown by leak features',fontproperties=proptease)
plt.setp(autotexts3, fontproperties=proptease)
plt.setp(texts3, fontproperties=proptease)



fig.subplots_adjust(wspace=1)
plt.show()
print 'number of leak features',fea[fea['is_leak']>0].shape[0],', number of non-leak features',fea[fea['is_leak']<1].shape[0]
```


####This figure shows that <font color='red'>31.8% </font> patients who have leak features in their records are also screeners and only <font color='red'>0.7%</font> patients who have leak features are not screeners. In other words, <font color='red'>97.8%</font> patients who have leak features are screeners. Since we didn't consider another leak case in this part, that a claim is present in procedure but not diagnosis, the actual patients with leak features could be more than the figure showed.

####Out of 404 important features <font color='red'>19.6% (79) </font> are leak features, out of which <font color='red'>97.5%</font> leak features indicate screeners.


####Next, let's find out what code pairs indicate strong leak.


```python
leak=fea[fea.is_leak>0].sort(columns='frequency',axis=0,ascending=False)
leak_single=leak[leak['feature'].apply(lambda x:x[:-1]).isin(fea['feature'])]
leak_single['single_feature']=leak_single['feature'].apply(lambda x:x[:-1])
fea['single_feature']=fea['feature']
fea['single_screen_rate']=fea['screen_rate']
fea['single_frequency']=fea['frequency']
leak_single=pd.merge(leak_single,fea[['single_screen_rate','single_feature','single_frequency']],on='single_feature',how='left')
print leak.shape, leak_single.shape
```

```python
#leak_single.head()
```

```python
N = 10
leak_sr = leak_single.screen_rate[:N]# 
single_sr = leak_single.single_screen_rate[:N]#
xlabelname= leak_single.single_feature[:N]
ind = np.arange(N)  # the x locations for the groups
width = 0.4       # the width of the bars

fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(311)

#fig, ax = plt.subplots()

rects1 = ax.bar(ind, leak_sr, width,  color='#FFBF00')#color='r')
rects2 = ax.bar(ind+width, single_sr, width,  color='c')
ax.set_ylabel('Screener Rate',fontweight='bold')
ax.set_title('Screener Rate of Top 10 Frequent Leak Features',fontweight='bold')
ax.set_xticks(ind+width)
ax.set_xticklabels(xlabelname,fontweight='bold')
plt.grid()
ax.set_ylim(0.,1.10)
ax.legend((rects1[0], rects2[0]), ('diagnosis_code + missing procedure_code (leak)', 'just diagnosis_code'),loc='lower left')
def autolabel(rects,ax):
    for c,rect in enumerate(rects):
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                '%.3f' % float(height),
                ha='center', va='bottom',fontweight='bold')
autolabel(rects1,ax)
autolabel(rects2,ax)

ax1 = fig.add_subplot(312)
leak_fr = leak_single.frequency[:N]# 
single_fr = leak_single.single_frequency[:N]

rects1 = ax1.bar(ind, leak_fr, width,  color='#FFBF00')#color='r')
rects2 = ax1.bar(ind+width, single_fr, width,  color='c')
ax1.set_ylabel('Frequency',fontweight='bold')
ax1.set_title('Frequency of Top 10 Frequent Leak Features',fontweight='bold')
ax1.set_xticks(ind+width)
ax1.set_xticklabels(xlabelname,fontweight='bold')
plt.grid()
#ax.set_ylim(0.,1.10)
ax1.legend((rects1[0], rects2[0]), ('diagnosis_code + missing procedure_code (leak)', 'just diagnosis_code'),loc='upper right')

autolabel(rects1,ax1)
autolabel(rects2,ax1)

ax2 = fig.add_subplot(313)
ax2.scatter(leak_single.frequency,leak_single.screen_rate,color='#FFBF00',marker='D',s=20,label='diagnosis_code + missing procedure_code (leak)')
ax2.scatter(leak_single.single_frequency,leak_single.single_screen_rate,color='c',label='just diagnosis_code',s=20)
ax2.legend(loc='upper right')
plt.xlim(-0.01,0.5)
plt.ylim(0.5,1.1)
ax2.set_title('Frequency vs Screener Rate for 46 leak features',fontweight='bold')
ax2.set_ylabel('Screener Rate',fontweight='bold')
ax2.set_xlabel('Frequency',fontweight='bold')
plt.grid()

fig.subplots_adjust(hspace=0.4)
plt.show()
print 'mean screener rate for leak features',leak_single.screen_rate.mean(),'summed frequency of leakfeatures',leak_single.frequency.sum()
```

####These figures show that the leak features strongly indicate screeners, and their predicting power are much stronger than the occurrence of the same diagnosis code without considering the missing pair (labelled as "just diagnosis code" in the figure). Please note that the green bar includes the yellow bar for each code. For example, <font color='red'>23%</font> patients have the leak feature with diagnosis code V72.31, and <font color='red'>99.8%</font> of them are screeners! For all 76 leak features, the average screener rate is <font color='red'>97.5%</font> while non-leak features has the average screener rate 84%. Except for diagnosis code V72.31, other leak features are not fequent, smaller than 3%.
